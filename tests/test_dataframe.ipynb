```python
import pandas as pd
from pandas import testing as tm
df = pd.read_pickle("./data.pkl")
df.head()
```

```python
indexed_df = df.set_index(['Onderwijssoort'])
indexed_df.info()
```

```python
def reset_known_index(df, column_name):
    reset_index = df.set_index([column_name])
    return reset_index

test_reset_known_index = reset_known_index(df, 'Onderwijssoort')
test_reset_known_index.info()
```

```python
test_reset_known_index.index
```

```python
tm.assert_index_equal(test_reset_known_index.index, indexed_df.index)
```

```python
changed_dtype_df = indexed_df.astype("int64")
changed_dtype_df.info()
```

```python
def change_dtype_columns(df, dtype):
    changed_df = df.astype(dtype)
    return changed_df

test_change_dtype_columns = change_dtype_columns(indexed_df, "int64")
test_change_dtype_columns.info()
```

```python
tm.assert_frame_equal(test_change_dtype_columns, changed_dtype_df)
```
